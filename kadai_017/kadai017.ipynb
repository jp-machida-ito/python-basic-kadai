```python
#ためし
class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age > 20:
      print(f"{self.name}は大人です")

    else:
      print(f"{self.name}は大人ではありません")

user1 = Human("太郎",10)
user2 = Human("花子",25)
user3 = Human("一郎",18)
user4 = Human("二郎",35)

users_list = [user1,user2,user3,user4]

for i in users_list:
  i.check_adult()


```
